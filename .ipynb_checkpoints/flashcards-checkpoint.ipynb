```python
with open('all_sent.csv', 'r', encoding = 'utf-8-sig', newline='') as readFile:
    reader = csv.reader(readFile, delimiter='\t')
    lines = list(reader)
    
cn_set = [lines[i][0] for i in range(len(lines))]

sent_level = np.zeros(len(lines))
hsk_chars = np.zeros(1)

for i in range(1, 7):
    
    if i == 1:
        with open('HSK1.txt', 'r', encoding = 'utf-8-sig') as readFile:
            reader = csv.reader(readFile)
            hsk_chars = list(reader)   
    else:
        with open(('HSK' + str(i) + '.txt'), 'r', encoding = 'utf-8-sig') as readFile:
            reader = csv.reader(readFile)
            hsk_chars = np.append(hsk_chars, list(reader))

for i in range(len(sent_level)):
    
    snt = cn_set[i]
    
    for j in range(len(snt)):
        if snt[j] in hsk_chars[0]:
            sent_level[i] = sent_level[i] + 1
        elif snt[j] in hsk_chars[1]:
            sent_level[i] = sent_level[i] + 2
        elif snt[j] in hsk_chars[2]:
            sent_level[i] = sent_level[i] + 3
        elif snt[j] in hsk_chars[3]:
            sent_level[i] = sent_level[i] + 4
        elif snt[j] in hsk_chars[4]:
            sent_level[i] = sent_level[i] + 5
        elif snt[j] in hsk_chars[5]:
            sent_level[i] = sent_level[i] + 6
        else:
            sent_level[i] = sent_level[i] + 0
            
    sent_level[i] = sent_level[i] / len(snt)
    
print(' -- sentence levels done -- ')
```

```python
for i in range(len(lines)): lines[i].append(sent_level[i])
```

```python
def set_tone(letter, tone):
    
    if letter == 'a':
        if tone == 1:
            mod = 'a' + u'\u0304'
        elif tone == 2:
            mod = 'a' + u'\u0301'
        elif tone == 3:
            mod = 'a' + u'\u030C'
        elif tone == 4:
            mod = 'a' + u'\u0300'
        elif tone == 5:
            mod = 'a'
    elif letter == 'e':
        if tone == 1:
            mod = 'e' + u'\u0304'
        elif tone == 2:
            mod = 'e' + u'\u0301'
        elif tone == 3:
            mod = 'e' + u'\u030C'
        elif tone == 4:
            mod = 'e' + u'\u0300'
        elif tone == 5:
            mod = 'e'
    elif letter == 'i':
        if tone == 1:
            mod = 'i' + u'\u0304'
        elif tone == 2:
            mod = 'i' + u'\u0301'
        elif tone == 3:
            mod = 'i' + u'\u030C'
        elif tone == 4:
            mod = 'i' + u'\u0300'
        elif tone == 5:
            mod = 'i'
    elif letter == 'o':
        if tone == 1:
            mod = 'o' + u'\u0304'
        elif tone == 2:
            mod = 'o' + u'\u0301'
        elif tone == 3:
            mod = 'o' + u'\u030C'
        elif tone == 4:
            mod = 'o' + u'\u0300'
        elif tone == 5:
            mod = 'o'
    elif letter == 'u':
        if tone == 1:
            mod = 'u' + u'\u0304'
        elif tone == 2:
            mod = 'u' + u'\u0301'
        elif tone == 3:
            mod = 'u' + u'\u030C'
        elif tone == 4:
            mod = 'u' + u'\u0300'
        elif tone == 5:
            mod = 'u'
        
    return mod
        
        
def replace_at_index(orig_str, ind, replcmt):
    
    new_str = orig_str
    
    if ind < len(orig_str):
        new_str = orig_str[0:ind] + replcmt + orig_str[ind + 1:]
        
    return new_str


def get_letter(fragment, vowel_ind, vowels, tone):
    
    if 'a' in vowels:
        fragment = replace_at_index(fragment, vowel_ind[vowels.index('a')], set_tone('a', tone))
        
    elif 'e' in vowels:
        fragment = replace_at_index(fragment, vowel_ind[vowels.index('e')], set_tone('e', tone))
        
    elif ('o' in vowels) and ('e' in vowels):
        fragment = replace_at_index(fragment, vowel_ind[vowels.index('o')], set_tone('o', tone))
        
    else:
        fragment = replace_at_index(fragment, vowel_ind[-1], set_tone(vowels[-1], tone))
        
    return fragment


def start_mod(ex):
    num = ['1', '2', '3', '4', '5']
    vwl = ['a', 'e', 'i', 'o', 'u']
    spt = ex.split()
    for i in range(len(spt)):

        vowel_ind = []
        vowels = []
        tone = 0
        addit = 0

        j = 0

        while j < (len(spt[i])):

            if (spt[i][j] not in vwl) and (spt[i][j] not in num):
                if (j + 1) != len(spt[i]):
                    if (spt[i][j + 1] not in num):            
                        vowel_ind = []
                        vowels = []
                        tone = 0

            elif spt[i][j] in vwl:
                vowel_ind.append(j)
                vowels.append(spt[i][j])

            elif spt[i][j] in num:

                if len(vowel_ind) != 0:
                    tone = int(spt[i][j])
                    spt[i] = get_letter(spt[i], vowel_ind, vowels, tone)

                    vowel_ind = []
                    vowels = []
                    tone = 0

                    addit += 1

            j += 1

            if j == 100:
                break
                
    fnl = (' ').join(spt)
    return fnl
```

```python
with open('all_sent.csv', 'r', encoding='utf-8-sig') as readFile:
    reader = csv.reader(readFile, delimiter='\t')
    lines  = list(reader)
```
